```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("DATASET_STROKES_BALANCEADO.csv",sep=";")
dataframe = dataframe.dropna()
clasificadores = ["no", "yes"]
```

```python
dataframe.head(10)
dataframe.describe()
dataframe.groupby("Stroke").size()
```

```python
import seaborn as sb
sb.catplot(x = "Stroke",data=dataframe,kind="count",height=2.5, aspect=.8)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
for column in dataframe.select_dtypes(include=['object']).columns:
    dataframe[column] = label_encoder.fit_transform(dataframe[column])
```

```python
y = dataframe["Stroke"]
x = dataframe.drop("Stroke", axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_mm = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_mm.fit(x_train, y_train)
```

```python
predicciones = knn_mm.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
tn, fp, fn, tp = confusion_matrix(y_test,predicciones).ravel()
```

```python
print(tn, fp, fn, tp)
```

```python
sns.heatmap(confusion_matrix(y_test,predicciones), annot=True, fmt="d")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
```

```python
print(classification_report(y_test,predicciones,target_names=clasificadores))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

y = dataframe["Stroke"]
x = dataframe.drop("Stroke", axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(x_train, y_train)
```

```python
pred_rf = modelo_rf.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print(classification_report(y_test, pred_rf, target_names=["no", "yes"]))

sns.heatmap(confusion_matrix(y_test, pred_rf), annot=True, fmt="d")
plt.title("Matriz de Confusión - Random Forest")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()
```

```python
tasa_error = []

for i in range(30, 100):
    modelo_rf = RandomForestClassifier(n_estimators=i, random_state=42)
    modelo_rf.fit(x_train, y_train)
    prediccion_i = modelo_rf.predict(x_test)
    tasa_error.append(np.mean(prediccion_i != y_test))
```

```python
valores = range(30, 100)
plt.plot(valores, tasa_error, color="green", marker="o", markerfacecolor="red", markersize="8")
```

```python
modelo_rf = RandomForestClassifier(n_estimators=94, random_state=42)
modelo_rf.fit(x_train, y_train)
```

```python
predicciones = modelo_rf.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, predicciones), annot=True, fmt="d")
plt.title("Matriz de Confusión - Random Forest (94 árboles)")
plt.xlabel("Predicción")
plt.ylabel("Valor Real")
plt.show()

clasificadores = ["no", "yes"]
print(classification_report(y_test, predicciones, target_names=clasificadores))
```
